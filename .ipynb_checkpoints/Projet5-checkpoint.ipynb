```python
data = Df_full_info_orders.dropna()
Df=Df.dropna()
info_data_unique(Df)
```

```python
cols_to_encode = Df.columns[(Df.dtypes=='datetime64[ns]')|(Df.dtypes=='string')|(Df.dtypes=='O')]

laben = preprocessing.LabelEncoder()

#encodons les valeurs non num√©riques
for encode in cols_to_encode :
    Df[encode] = laben.fit_transform(Df[encode])
Df.shape
```

```python
X = Df.values
X.shape
```

```python
from sklearn import preprocessing

std_scale = preprocessing.StandardScaler().fit(X)
X_scaled = std_scale.transform(X)
```

```python
from sklearn import decomposition
n_comp = X.shape[1]
print(n_comp)
pca = decomposition.PCA(n_components=n_comp)
pca.fit(X_scaled)
```

```python
print(pca.explained_variance_ratio_.cumsum())
```

```python
plt.plot(list(range(1,n_comp+1)),pca.explained_variance_ratio_.cumsum())
plt.title('inertie par features')
plt.show()
```

```python
# 1. Choisissons le clustering avec kmeans 
from sklearn import cluster
myclust = cluster.KMeans(n_clusters=2)
myclust.fit(X)
```

```python
# 2. Visualisation
# On commence par r√©duire la dimension des donn√©es avec tSNE. On scale d‚Äôabord les donn√©es :
from sklearn import preprocessing
std_scale = preprocessing.StandardScaler().fit(X)
X_scaled = std_scale.transform(X)
```

```python
# Puis on applique tSNE aux donn√©es scal√©es :
from sklearn import manifold
tsne = manifold.TSNE(n_components=2, init='pca')
X_trans = tsne.fit_transform(X_scaled)
```

```python
y= data.review_score.values
```

```python
# Puis on visualise, selon la m√©thode donn√©e dans le coup de pouce :
from matplotlib import pyplot as plt # Dans Jupyter : % pylab inline
plt.figure(figsize=(15,15))
for i in range(X.shape[0]): # pour chaque point
    plt.text(X_trans[i, 0], X_trans[i, 1], str('%s' % y[i]),
            color=plt.cm.Set2(myclust.labels_[i]/10.))
    
# On red√©finit les axes pour bien voir tous les points
plt.xlim([np.min(X_trans[:, 0]), np.max(X_trans[:, 0])])
plt.ylim([np.min(X_trans[:, 1]), np.max(X_trans[:, 1])])

# On affiche l'image
plt.show()
```

```python
# projeter X sur les composantes principales
X_projected = pca.transform(X_scaled)

# afficher chaque observation
plt.scatter(X_projected[:, 0], X_projected[:, 1],
    # colorer en utilisant la variable 'Rank'
    c=data.get('review_score'))

plt.xlim([-5.5, 5.5])
plt.ylim([-4, 4])
plt.colorbar()
```
